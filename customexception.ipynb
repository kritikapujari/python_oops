Custom Exception(Raise and Throw an Exception)

```python
class Error(Exception):
    pass
class dobException(Error):
    pass
```

```python
year=int(input("Enter the dob"))
age = 2024-year
try:
    if age<=30 and age>=20:
        print("age valid.You can apply for the exam")
    else:
        raise dobException
except dobException:
    print("Sorry age should be greater than 20 and less than 30")
```
